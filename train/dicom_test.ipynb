```python
nifti_1 = nib.load("C:/Users/user/Desktop/인턴/I001.nii.gz")
nifti_2 = nib.load("C:/Users/user/Desktop/인턴/I002.nii.gz")

nifti_image_1 = nifti_1.get_fdata()
nifti_image_2 = nifti_2.get_fdata()

print(type(nifti_image_1), nifti_image_1.shape, nifti_image_1.dtype, np.min(nifti_image_1), np.max(nifti_image_1))
print(type(nifti_image_2), nifti_image_2.shape, nifti_image_2.dtype, np.min(nifti_image_2), np.max(nifti_image_2))
```

```python
slice1_0 = nifti_image_1[192, :, :]
slice1_1 = nifti_image_1[:, 192, :]
slice1_2 = nifti_image_1[:, :, 16]

slice2_0 = nifti_image_2[192, :, :]
slice2_1 = nifti_image_2[:, 192, :]
slice2_2 = nifti_image_2[:, :, 16]
```

```python
def show_slices(slices):
    fig, axes = plt.subplots(1, len(slices))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice.T, cmap = "gray", origin = "lower")
```

```python
show_slices([slice1_0, slice1_1, slice1_2])
```

```python
show_slices([slice2_0, slice2_1, slice2_2])
```
